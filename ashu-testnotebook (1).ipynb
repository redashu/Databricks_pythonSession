```python
import builtins

class NumberOperations:

    def __init__(self):
        self.user_list = []

    # Take input from user
    def take_input(self):
        for i in range(7):
            user_input = builtins.input("Enter number (q to quit): ")

            if user_input == 'q':
                break
            else:
                self.user_list.append(int(user_input))

    # Find sum
    def sum_number(self):
        total = 0

        for i in self.user_list:
            total += i

        return total

    # Sort list
    def sort_numbers(self):
        self.user_list.sort()
        return self.user_list


# Create object
obj = NumberOperations()

# Call methods
obj.take_input()

print("User List:", obj.user_list)
print("Sum:", obj.sum_number())
print("Sorted List:", obj.sort_numbers())
```
