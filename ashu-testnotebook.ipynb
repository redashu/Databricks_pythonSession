```python
# tuple 
t1 = (3,5,7,88,2,100)
t2 = (5,6.7,"Hello Databricks",100)
```

```python
print(type(t1),type(t2))
# check len of tuple
print(len(t1),len(t2))
#
print(t1[-1])
t2[2].split()
```

```python
t3 = t1 + t2
print(t3)
# we can't add  , remove , append elements in tuple as it is immutable
```
